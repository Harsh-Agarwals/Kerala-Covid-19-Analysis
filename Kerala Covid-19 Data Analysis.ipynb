### Checking for duplicate columns

```python
covid_df.duplicated().sum()
```

### Changing the datetime column

```python
pd.to_datetime(covid_df.Date)
```

Adding 3 new columns named total_cases, total_deaths and total_recovered using the .cumsum() or cummulative sum method. Each column will add up the last sum and the current value of the column.

```python
covid_df['total_cases'] = covid_df.Confirmed.cumsum()
covid_df['total_deaths'] = covid_df.Decreased.cumsum()
covid_df['total_recovered'] = covid_df.Recovered.cumsum()
```

```python
covid_df
```

In the covid_df dataset, any column of total_cases, total_deaths and total_recovered points to the total cases, deaths and recovered population upto that date.

## Exploratory Analysis and Visualization

Visualising the data using matplotlib and seaborn library. 
And analysing the overall result.

First step involves importing the matplotlib and seaborn library

Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

### Visualizing new covid cases

Plotting this graph using plt.plot() method and providing the method with x and y data for the plot and adding the design for the plot at last.

```python
plt.figure(figsize=(16, 8))
plt.plot(covid_df.Date, covid_df.Confirmed, 'o-r')
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.title("COVID-19 Kerala confirmed cases")
plt.xticks(rotation='vertical');
```

Now, plotting cases per day using plt.bar method. this will plot the bar graph. the method contains x and y part for the columns. 

```python
# Plotting the bar graph

plt.figure(figsize=(12, 6))
plt.bar(covid_df.Date, covid_df.Confirmed, color='maroon')
plt.xlabel('Date')
plt.ylabel('Day Confirmed')
plt.title('Cases Per Day');
```

Getting the graph from 450th day i.e. data of about last 6 month

```python
plt.figure(figsize=(12, 6))
plt.plot(covid_df.loc[450:].Date, covid_df.loc[450:].Confirmed, '-b')
plt.xlabel('Date')
plt.ylabel('Day Confirmed')
plt.title('Cases Per Day');
```

```python
plt.figure(figsize=(12, 6))
plt.plot(covid_df.loc[::10].Date, covid_df.loc[::10].Confirmed, '-b')
plt.xlabel('Date')
plt.ylabel('Day Confirmed')
plt.title('Cases Per Day');
```

From the graph, it can be seen that 2nd peak is attained at 460th day from beginning and from 410th day, this peak started.

Current Situation - cases are increasing and new peak can be visible

```python
plt.plot(covid_df.Date, covid_df.Confirmed, '-r')
plt.xlabel('Days')
plt.ylabel('Confirmed Cases')
plt.title('Cases Per day');
```

## Visualizing deaths

Using .plot method to plot the graph and providing it with data of date for x axis and decreased column to the y axis.

```python
plt.plot(covid_df[400:].Date, covid_df[400:].Decreased)
plt.xlabel("Dates")
plt.ylabel("Deaths")
plt.title("Covid-19 Deaths");
```

Using DatetimeIndex of the pandas method to get the column of month and year from the date provided. 

```python
covid_df['Month'] = pd.DatetimeIndex(covid_df.Date).month
covid_df['year'] = pd.DatetimeIndex(covid_df.Date).year
```

```python
covid_df.sample(10)
```

.groupby method is used to group the dataset based on the column provided. usign .groupby() method on month and year column to group dataset on month and year basis. the final grouped data we get is for each year divided on the monthly basis.

```python
cov_month_wise = covid_df.groupby(['Month', 'year'])[['Confirmed', 'Recovered', 'Decreased']].sum()
```

```python
cov_month_wise
```

In the grouped dataset, we can see that the dataset is not perfectly indexed. So, resetting index using reset_index method of the pandas.

```python
cov_month_wise.reset_index("Month", inplace=True)
cov_month_wise.reset_index("year", inplace=True)
```

```python
cov_month_wise
```

Since 2021 Sept had just started, we will drop 9th month of dataset of 2021

```python
cov_month_wise.drop(17, inplace=True)
```

```python
cov_month_wise.reset_index(inplace=True)
```

```python
cov_month_wise
```

Now stacking the data based on year.

```python
cov_month_wise[cov_month_wise.year == 2020]
```

```python
cov_month_wise.columns
```

Now comparing the covid cases of both years using graph. we will use plt.plot method for plotting the graph.

```python
plt.plot(cov_month_wise[cov_month_wise.year == 2020].Month, cov_month_wise[cov_month_wise.year == 2020].Confirmed, 'o-r')
plt.plot(cov_month_wise[cov_month_wise.year == 2021].Month, cov_month_wise[cov_month_wise.year == 2021].Confirmed, 'o-b')
plt.legend(['2020', '2021'])
plt.xlabel('Month')
plt.ylabel('Confirmed Cases')
plt.title('Total Cases v/s Month');
```
